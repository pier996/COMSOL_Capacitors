```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter
from IPython.display import display, Markdown, Latex
from scipy.optimize import curve_fit
%matplotlib inline
from scipy.integrate import ode
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import HTML
```

```python
#loading data
data = np.loadtxt("data/angle/volt-theta.txt")

#putting data in lists
x = data[:,0]
y = data[:,1]



#creating the graph and setting the dimensions
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_xlim(0, 60)
ax.set_ylim(-0, 0.04)

# function to define minor tick, if we want something written under the minor tick we have to write it in the first return
def minor_tick(x, pos):
    if not x % 1.0:
        return ""
    return "%.2f" % x

#tick numbers and positions
#ax.xaxis.set_major_locator(MultipleLocator(1.000))
ax.xaxis.set_minor_locator(AutoMinorLocator(2))
#ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.xaxis.set_minor_formatter(FuncFormatter(minor_tick))

#tick characteristics
ax.tick_params(which='major', width=1.0, labelsize = 11)
ax.tick_params(which='major', length=10)
ax.tick_params(which='minor', width=1.0, labelsize=10)
ax.tick_params(which='minor', length=5, labelsize=10, labelcolor='0.25')

#grid style
ax.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)

#plotting point
ax.plot(x, y, linewidth=0,
        marker='o', markerfacecolor='r', markeredgecolor='b', markersize=5)
#ax.scatter(x=x, y=y, marker='o', markeredgecolor='k', label="COMSOL")
#ax.plot(x, y, linewidth=1,  c="cornflowerblue", label="COMSOL")

#errors
#err = np.abs(y)*(0.0000361)
#ax.errorbar(x, y, xerr = 0 , yerr = err, fmt='.',capthick=1, capsize=10, c='k')

#plot a line, Y (and maybe also X) are not necessary
#X = np.linspace(0, 1600, 100)
#Y = np.linspace(0, 1,100)
#ax.plot(X, 1+Y*0, linewidth=1, c="red")


# labels and title
ax.set_title('Capacitor 1',weight='bold', size=24)
ax.set_xlabel('Voltage (V)', fontsize=20)
ax.set_ylabel('Deflection Angle (mrad)', fontsize=20)

#fit
#m, b = np.polyfit(x, y, 1)
#print(m,b)
#plt.plot(x, m*x + b, c = 'y', label="linear fit")

#legend
#ax.legend(loc='upper left', fontsize=18)

plt.show()


```
